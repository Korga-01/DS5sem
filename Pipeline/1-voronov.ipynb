```python
import pandas as pd

train_data = pd.read_csv('C:/Users/sahav/Desktop/Data Science/Pipeline/train.csv')
test_data = pd.read_csv('C:/Users/sahav/Desktop/Data Science/Pipeline/test.csv')


train_data.head(15)
```

14 столбцов.

**PassengerId**
Описание: Уникальный идентификатор пассажира.
Тип данных: Строковый.

**HomePlanet**
Описание: Планета, с которой прибыл пассажир (например, Земля, Европа, Марс).
Тип данных: Категориальный.

**CryoSleep**
Описание: Находился ли пассажир в состоянии "криосна" во время путешествия (булевый признак: True или False).
Тип данных: Булевый (логический).

**Cabin**
Описание: Номер каюты пассажира, который состоит из трех частей: палуба (Deck), номер (Num) и сторона (Side) (например, B/45/P — палуба B, каюта 45, правая сторона).
Тип данных: Строковый.

**Destination**
Описание: Пункт назначения пассажира (например, TRAPPIST-1e, 55 Cancri e, PSO J318.5-22).
Тип данных: Категориальный.

**Age**
Описание: Возраст пассажира.
Тип данных: Числовой.

**VIP**
Описание: Был ли пассажир VIP (булевый признак: True или False).
Тип данных: Булевый (логический).

**RoomService**
Описание: Расходы пассажира на обслуживание в номере (в денежных единицах).
Тип данных: Числовой.

**FoodCourt**
Описание: Расходы пассажира в фудкорте на борту.
Тип данных: Числовой.

**ShoppingMall**
Описание: Расходы пассажира в торговом центре на борту.
Тип данных: Числовой.

**Spa**
Описание: Расходы пассажира на посещение спа-салона на борту.
Тип данных: Числовой.

**VRDeck**
Описание: Расходы пассажира на использование VR-деков (виртуальной реальности) на борту.
Тип данных: Числовой.

**Name**
Описание: Имя пассажира.
Тип данных: Строковый.

**Transported**
Описание: Целевой признак — был ли пассажир перемещен на другой корабль (True или False).
Тип данных: Булевый (логический).

много незаполненных данных

```python
# Замена пропусков модой для категориальных признаков
train_data['HomePlanet'] = train_data['HomePlanet'].fillna(train_data['HomePlanet'].mode()[0])
train_data['CryoSleep'] = train_data['CryoSleep'].fillna(False).astype(bool)
train_data['Destination'] = train_data['Destination'].fillna(train_data['Destination'].mode()[0])
train_data['VIP'] = train_data['VIP'].fillna(False).astype(bool)

# Замена пропусков медианой для числовых признаков
train_data['Age'] = train_data['Age'].fillna(train_data['Age'].median())
train_data['RoomService'] = train_data['RoomService'].fillna(0)
train_data['FoodCourt'] = train_data['FoodCourt'].fillna(0)
train_data['ShoppingMall'] = train_data['ShoppingMall'].fillna(0)
train_data['Spa'] = train_data['Spa'].fillna(0)
train_data['VRDeck'] = train_data['VRDeck'].fillna(0)

# Замена пропусков для Cabin
train_data['Cabin'] = train_data['Cabin'].fillna('Unknown')

# Удаление ненужных колонок
train_data = train_data.drop(columns=['Name', 'PassengerId'])

```

Теперь добавим новый признак.

AgeBand

```python
# Функция для назначения возрастных групп
def age_band(age):
    if pd.isnull(age):  # Проверяем на NaN
        return 'Unknown'
    elif age <= 12:
        return 0
    elif age <= 19:
        return 1
    elif age <= 35:
        return 2
    elif age <= 50:
        return 3
    elif age <= 65:
        return 4
    else:
        return 5

# Применяем функцию к столбцу Age для создания нового столбца AgeBand
train_data['AgeBand'] = train_data['Age'].apply(age_band)
```
